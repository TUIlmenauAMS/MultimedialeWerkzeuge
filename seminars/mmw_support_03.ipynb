```python
S = requests.Session()

URL = "https://api.lyrics.ovh/v1/slash/paradise%20city"

#PARAMS = ['slash','paradise%20city']

R = S.get(url=URL)
R.json()
```

```python
from requests.sessions import Request
S = requests.Session()

URL = "https://www.songsterr.com/a/ra/songs/byartists.json"

PARAMS = {'artists':'Metallica'}

R = S.get(url=URL, params=PARAMS)
R.json()
```
